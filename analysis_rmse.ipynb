```python
vars_to_predict = ['tas', 'dtr', 'pr', 'pr90']
```

```python
def print_min_rmses(dense_1M, dense_10M, rand_dense_1M, rand_dense_10M):
  for var_idx in range(4):
    # Best standard network performances for each layer depth at both parameter counts
    dense_best_perfs_1M = [min(dense_1M[i][var_idx]) for i in range(dense_1M.shape[0])]
    dense_best_perfs_10M = [min(dense_10M[i][var_idx]) for i in range(dense_10M.shape[0])]
    dense_best_perfs_all = dense_best_perfs_1M + dense_best_perfs_10M

    if np.argmin(dense_best_perfs_all) // dense_1M.shape[0] == 0:
      dense_best = dense_1M
      dense_best_params = '1M params'
    else:
      dense_best = dense_10M
      dense_best_params = '10M params'

    dense_best_layer_count = np.argmin(dense_best_perfs_all) % dense_1M.shape[0]
    dense_best_idx = np.argmin(dense_best[dense_best_layer_count][var_idx])

    # Best RandDense network performances for each layer depth at both parameter counts
    rand_dense_best_perfs_1M = [min(rand_dense_1M[i][var_idx]) for i in range(rand_dense_1M.shape[0])]
    rand_dense_best_perfs_10M = [min(rand_dense_10M[i][var_idx]) for i in range(rand_dense_10M.shape[0])]
    rand_dense_best_perfs_all = rand_dense_best_perfs_1M + rand_dense_best_perfs_10M

    if np.argmin(rand_dense_best_perfs_all) // rand_dense_1M.shape[0] == 0:
      rand_dense_best = rand_dense_1M
      rand_dense_best_params = '1M params'
    else:
      rand_dense_best = rand_dense_10M
      rand_dense_best_params = '10M params'

    rand_dense_best_layer_count = np.argmin(rand_dense_best_perfs_all) % rand_dense_1M.shape[0]
    rand_dense_best_idx = np.argmin(rand_dense_best[rand_dense_best_layer_count][var_idx])

    print(vars_to_predict[var_idx])
    print(f'\tStandard:  {round(min(dense_best_perfs_all),3)}\t({dense_best_params}, {dense_best_layer_count+2} Layers, Model {dense_best_idx})')
    print(f'\tRandDense: {round(min(rand_dense_best_perfs_all),3)}\t({rand_dense_best_params}, {rand_dense_best_layer_count+2} Layers, Model {rand_dense_best_idx})')
```

```python
# Best RMSES across all generated Dense and RandDense networks
print_min_rmses(dense_1M_total, dense_10M_total, rand_dense_1M_total, rand_dense_10M_total)
```

```python
# Best RMSES across all generated CNN and CNN RandDense networks
print_min_rmses(cnn_1M_total, cnn_10M_total, cnn_rand_dense_1M_total, cnn_rand_dense_10M_total)
```

```python
# Best RMSES across all generated CNN-LSTM and CNN-LSTM RandDense networks
print_min_rmses(cnn_lstm_1M_total, cnn_lstm_10M_total, cnn_lstm_rand_dense_1M_total, cnn_lstm_rand_dense_10M_total)
```
